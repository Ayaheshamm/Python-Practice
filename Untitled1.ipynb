```python
!pip install mlxtend -q

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
from google.colab import files
uploaded = files.upload()  # upload Egyptian_Grocery_Translated.xlsx

df = pd.read_excel(list(uploaded.keys())[0], header=None)

# split each row into a list of items
transactions = [row[0].split(', ') for _, row in df.iterrows()]
```

```python
te = TransactionEncoder()
df_encoded = pd.DataFrame(te.fit_transform(transactions), columns=te.columns_)
```

```python
# items appearing in at least 10% of transactions
frequent_itemsets = apriori(df_encoded, min_support=0.1, use_colnames=True)
frequent_itemsets.sort_values('support', ascending=False)
```

```python
# keep rules where confidence >= 0.3
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.3)
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules.sort_values('confidence', ascending=False)
```

```python
strong = rules[(rules['confidence'] >= 0.6) & (rules['lift'] > 1)]
weak   = rules[rules['confidence'] < 0.4]

print("Strong Rules:")
print(strong.head(2).to_string(index=False))
print("\nWeak Rules:")
print(weak.head(2).to_string(index=False))
```

```python
print(f"Transactions:      {len(transactions)}")
print(f"Frequent itemsets: {len(frequent_itemsets)}")
print(f"Rules generated:   {len(rules)}")
print(f"Strong rules:      {len(strong)}")
print(f"Weak rules:        {len(weak)}")

print("""
Key Insights:
- Pepsi, Eggs, Lentils, Sugar, Oil are the most frequent  everyday Egyptian staples.
- Strong rules reflect real habits e.g. Koshari ingredients bought together.
- Weak rules are mostly coincidental with low predictive value.
- Strong rules can guide shelf placement or bundle promotions.
""")
```
